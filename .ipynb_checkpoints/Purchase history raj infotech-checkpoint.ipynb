```python
# creating a new data frame with only two cols
df2 = df.iloc[:,:2]
```

```python
i = 0
while i < len(df):
    date = df2.iloc[i, 0]
    particulars = df2.iloc[i, 1]
    i += 1

    # Check if the next row has a null date, and if so, add the product to the 'products' column in df2
    while i < len(df) and pd.isna(df.iloc[i, 0]):
        df2.loc[i, 'Date'] = date
        df2.loc[i, 'Particulars'] = particulars
        df2.loc[i, 'products'] = df.iloc[i, 1]
        i += 1

    # Check if 'Grand Total' is found in the 'Particulars' column and break the loop if it is
    if df2.loc[i - 1, 'Particulars'] == "Grand Total":
        break
```
